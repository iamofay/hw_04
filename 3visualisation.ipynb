```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('tips')
print(df)
```

```python
# # Постройте гистограмму распределения признака 'total_bill'
sns.displot(df['total_bill'])
plt.show()
```

```python
# # Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

sns.scatterplot(data=df, x='total_bill', y='tip')
plt.show()
```

```python
# # Постройте pairplot

sns.pairplot(df)
plt.show()
```

```python
# # Постройте график взаимосвязи между признаками 'total_bill' и 'day'

sns.scatterplot(data=df, x='total_bill', y='day')
plt.show()
```

```python
# # Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

grid = sns.FacetGrid(df, col='time')
grid.map(plt.hist, 'tip');
```

```python
# # Постройте два графика scatterplot,
# # представляющих взаимосвязь между признаками 'total_bill' и 'tip'
# # один для Male, другой для Female
# # и раскрасьте точки в зависимоти от признака 'smoker'

plt.subplot(1, 2, 1)
sns.scatterplot(data=df[((df['sex'] == 'Male') & (df['smoker'] == 'Yes'))], x='total_bill', y='tip', color='green')
sns.scatterplot(data=df[((df['sex'] == 'Male') & (df['smoker'] == 'No'))], x='total_bill', y='tip', color='red')
plt.subplot(1, 2, 2)
sns.scatterplot(data=df[((df['sex'] == 'Female') & (df['smoker'] == 'Yes'))], x='total_bill', y='tip', color='green')
sns.scatterplot(data=df[((df['sex'] == 'Female') & (df['smoker'] == 'No'))], x='total_bill', y='tip', color='red')
plt.show()
```

```python
# Сделайте выводы по анализу датасета и построенным графикам. По желанию можете продолжить анализ данных и также отразить это в выводах.

# Для визуализации выводов можно построить пару графиков. График зависимости мужчин и женщин с распределением по дням и график зависимости времени посещения по дням.

df.groupby('day', observed=True)['sex'].value_counts().unstack().plot(kind='bar')
df.groupby('day', observed=True)['time'].value_counts().unstack().plot(kind='bar')

# По графикам видно, что по выходным платят приемущественном мужчины, при этом не зависимо от пола посетители предпочитают ходить на ужин, в четверг на обед, а в пятницу посещаемость низкая по сравнению с другими днями
```

```python
# Так же построим график зависимости количества чаевых по дням
df.groupby('day', observed=True)['total_bill'].sum().plot(kind='bar')
```

```python
# И распределение счета в зависимости от пола
df.groupby('sex', observed=True)['total_bill'].mean().plot(kind='bar')
# И дополнительно убедимся в том, что самые прибыльные дни это выходные, а самый неприбыльный день, это пятница. При этом мужчины в среднем оплачивают более дорогие счета.
```

```python
df.groupby('sex', observed=True)['tip'].mean().plot(kind='bar')
# И платят больше чаевых
```
